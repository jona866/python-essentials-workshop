## Ejercicio 1: Variables y Print

**Enunciado:** Crea tres variables: `mi_nombre` con tu nombre, `mi_edad` con tu edad, `mi_hobby` con tu pasatiempo favorito. Luego imprime un mensaje que diga: "Me llamo [nombre], tengo [edad] a√±os y me gusta [hobby]"

```python
# Soluci√≥n
mi_nombre = "Juan"
mi_edad = 25
mi_hobby = "programar"

print(f"Me llamo {mi_nombre}, tengo {mi_edad} a√±os y me gusta {mi_hobby}")

# Explicaci√≥n:
# 1. Creamos tres variables con los datos solicitados
# 2. Usamos f-string (f"...") para formatear el mensaje
# 3. Las llaves {} insertan el valor de las variables en el texto
```

**Alternativas:**

```python
# Alternativa 1: Usando concatenaci√≥n con +
print("Me llamo " + mi_nombre + ", tengo " + str(mi_edad) + " a√±os y me gusta " + mi_hobby)

# Alternativa 2: Usando el m√©todo .format()
print("Me llamo {}, tengo {} a√±os y me gusta {}".format(mi_nombre, mi_edad, mi_hobby))

# Nota: La f-string es la forma m√°s moderna y legible (Python 3.6+)
```

## Ejercicio 2: Operaciones Matem√°ticas

**Enunciado:** Crea dos variables num√©ricas `a = 15` y `b = 4`, luego calcula e imprime: suma, resta, multiplicaci√≥n, divisi√≥n y m√≥dulo.

```python
# Soluci√≥n
a = 15
b = 4

suma = a + b
resta = a - b
multiplicacion = a * b
division = a / b
modulo = a % b

print(f"Suma: {a} + {b} = {suma}")
print(f"Resta: {a} - {b} = {resta}")
print(f"Multiplicaci√≥n: {a} √ó {b} = {multiplicacion}")
print(f"Divisi√≥n: {a} √∑ {b} = {division}")
print(f"M√≥dulo: {a} % {b} = {modulo}")

# Explicaci√≥n:
# - El m√≥dulo (%) devuelve el residuo de la divisi√≥n
# - 15 % 4 = 3 porque 15 = 4√ó3 + 3
```

**Enunciado:** Pide al usuario dos n√∫meros (usa `input()`), convi√©rtelos a enteros y muestra su suma.

**Enunciado:** Pide un n√∫mero al usuario y determina si es par o impar.

```python
# Soluci√≥n
numero = int(input("Ingresa un n√∫mero: "))

if numero % 2 == 0:
    print(f"{numero} es par")
else:
    print(f"{numero} es impar")

# Explicaci√≥n:
# - Un n√∫mero es par si al dividirlo entre 2 el residuo es 0
# - El operador % (m√≥dulo) devuelve el residuo de una divisi√≥n
# - Ejemplos: 4 % 2 = 0 (par), 5 % 2 = 1 (impar)
```

## Ejercicio 7: Calificaciones

**Enunciado:** Pide una calificaci√≥n (0-100) y muestra el nivel correspondiente.

```python
# Soluci√≥n
calificacion = int(input("Ingresa tu calificaci√≥n (0-100): "))

if calificacion >= 90:
    print("Excelente")
elif calificacion >= 80:
    print("Muy bien")
elif calificacion >= 70:
    print("Bien")
elif calificacion >= 60:
    print("Suficiente")
else:
    print("Reprobado")

# Explicaci√≥n:
# - elif significa "else if" (sino, si...)
# - Python eval√∫a las condiciones en orden
# - Cuando una condici√≥n es True, ejecuta ese bloque y termina
# - Por eso no necesitamos verificar calificacion < 90 en el elif
```

## Ejercicio 8: Ciclo For B√°sico

**Enunciado:** Usa un ciclo `for` para imprimir los n√∫meros del 1 al 10.

```python
# Soluci√≥n
for numero in range(1, 11):
    print(numero)

# Explicaci√≥n:
# - range(1, 11) genera n√∫meros del 1 al 10 (no incluye el 11)
# - En cada iteraci√≥n, 'numero' toma el siguiente valor
# - El ciclo se repite hasta que se acaban los valores
```

```python
# Versi√≥n con m√°s formato:
for numero in range(1, 11):
    print(f"N√∫mero: {numero}")
```

## Ejercicio 9: Suma de N√∫meros

**Enunciado:** Calcula la suma de los n√∫meros del 1 al 100 usando un ciclo.

```python
# Soluci√≥n
suma = 0

for numero in range(1, 101):
    suma = suma + numero

print(f"La suma de los n√∫meros del 1 al 100 es: {suma}")

# Explicaci√≥n:
# 1. Inicializamos suma en 0
# 2. En cada iteraci√≥n sumamos el n√∫mero actual
# 3. suma = suma + numero se puede escribir tambi√©n como suma += numero
# 4. Al final, suma contiene la suma total
```

```python
# Versi√≥n con += (operador de asignaci√≥n compuesta):
suma = 0
for numero in range(1, 101):
    suma += numero
print(f"Suma total: {suma}")
```

## Ejercicio 10: Tabla de Multiplicar

**Enunciado:** Pide un n√∫mero al usuario e imprime su tabla de multiplicar del 1 al 10.

```python
# Soluci√≥n
numero = int(input("Ingresa un n√∫mero: "))

print(f"\nTabla de multiplicar del {numero}:")
print("-" * 25)

for i in range(1, 11):
    resultado = numero * i
    print(f"{numero} √ó {i} = {resultado}")

# Explicaci√≥n:
# - Usamos for para iterar del 1 al 10
# - En cada iteraci√≥n multiplicamos el n√∫mero por i
# - print("\n") agrega una l√≠nea en blanco al inicio
# - "-" * 25 crea una l√≠nea de 25 guiones
```

## Ejercicio 11: Contar Pares

**Enunciado:** Cuenta cu√°ntos n√∫meros pares hay entre 1 y 50.

```python
# Soluci√≥n
contador = 0

for numero in range(1, 51):
    if numero % 2 == 0:
        contador += 1

print(f"Hay {contador} n√∫meros pares entre 1 y 50")

# Explicaci√≥n:
# 1. Inicializamos contador en 0
# 2. Recorremos los n√∫meros del 1 al 50
# 3. Si el n√∫mero es par (residuo 0 al dividir entre 2), incrementamos contador
# 4. Al final imprimimos el total
```

```python
# Versi√≥n m√°s eficiente (saltando de 2 en 2):
contador = 0
for numero in range(2, 51, 2):  # Empieza en 2, va hasta 50, salta de 2 en 2
    contador += 1
print(f"Hay {contador} n√∫meros pares entre 1 y 50")

# O simplemente:
pares = len(range(2, 51, 2))
print(f"Hay {pares} n√∫meros pares entre 1 y 50")
```

## Ejercicio 13: Funci√≥n de √Årea

**Enunciado:** Crea una funci√≥n `area_rectangulo(base, altura)` que calcule y retorne el √°rea de un rect√°ngulo.

```python
# Soluci√≥n
def area_rectangulo(base, altura):
    area = base * altura
    return area

# Probar la funci√≥n
resultado = area_rectangulo(5, 3)
print(f"√Årea del rect√°ngulo: {resultado}")

print(f"√Årea de un rect√°ngulo 10√ó4: {area_rectangulo(10, 4)}")
print(f"√Årea de un rect√°ngulo 7√ó8: {area_rectangulo(7, 8)}")

# Explicaci√≥n:
# - La funci√≥n recibe dos par√°metros: base y altura
# - Calcula el √°rea (base √ó altura)
# - Retorna el resultado
# - √Årea del rect√°ngulo = base √ó altura
```

```python
# Versi√≥n m√°s compacta:
def area_rectangulo(base, altura):
    return base * altura
```

## Ejercicio 14: Funci√≥n Mayor de Dos

**Enunciado:** Crea una funci√≥n `mayor(a, b)` que retorne el mayor de dos n√∫meros.

```python
# Soluci√≥n
def mayor(a, b):
    if a > b:
        return a
    else:
        return b

# Probar la funci√≥n
print(f"El mayor entre 10 y 5 es: {mayor(10, 5)}")
print(f"El mayor entre 3 y 8 es: {mayor(3, 8)}")
print(f"El mayor entre 7 y 7 es: {mayor(7, 7)}")

# Explicaci√≥n:
# - Comparamos a y b
# - Si a es mayor, retornamos a
# - Si no, retornamos b (que incluye el caso de igualdad)
```

```python
# Versi√≥n usando operador ternario (m√°s compacta):
def mayor(a, b):
    return a if a > b else b

# O simplemente usando la funci√≥n built-in max():
def mayor(a, b):
    return max(a, b)
```

## Ejercicio 15: Iterar sobre Lista

**Enunciado:** Dada la lista `frutas = ["manzana", "banana", "naranja", "uva"]`, imprime cada fruta precedida por su n√∫mero de posici√≥n (empezando en 1).

```python
# Soluci√≥n 1: Usando enumerate()
frutas = ["manzana", "banana", "naranja", "uva"]

for indice, fruta in enumerate(frutas, start=1):
    print(f"{indice}. {fruta}")

# Explicaci√≥n:
# - enumerate() devuelve pares (√≠ndice, valor)
# - start=1 hace que el conteo empiece en 1 en lugar de 0
# - En cada iteraci√≥n obtenemos el n√∫mero y la fruta
```

```python
# Soluci√≥n 2: Usando range() y len()
frutas = ["manzana", "banana", "naranja", "uva"]

for i in range(len(frutas)):
    print(f"{i + 1}. {frutas[i]}")

# Explicaci√≥n:
# - len(frutas) devuelve 4 (cantidad de elementos)
# - range(4) genera [0, 1, 2, 3]
# - Sumamos 1 para que empiece en 1 en lugar de 0
# - Accedemos a cada fruta con frutas[i]
```

```python
# Soluci√≥n 3: Usando un contador manual
frutas = ["manzana", "banana", "naranja", "uva"]
contador = 1

for fruta in frutas:
    print(f"{contador}. {fruta}")
    contador += 1

# Nota: La primera soluci√≥n con enumerate() es la m√°s pyth√≥nica
```
