## Imprimir M√∫ltiples Variables

Podemos mostrar varias variables en una sola l√≠nea:

````python
nombre = "Carlos"
edad = 30
ciudad = "M√©rida"

print("Nombre:", nombre)
print("Edad:", edad)
print("Ciudad:", ciudad)
````

## Reglas para Nombrar Variables

‚úÖ **Permitido:**
- Comenzar con letra o gui√≥n bajo: `nombre`, `_total`
- Usar letras, n√∫meros y guiones bajos: `edad2`, `primer_nombre`
- Usar min√∫sculas con guiones bajos: `mi_variable`

‚ùå **No permitido:**
- Comenzar con n√∫meros: `2nombre`
- Usar espacios: `mi variable`
- Usar caracteres especiales: `mi-variable`, `mi@variable`
- Usar palabras reservadas: `if`, `for`, `while`

````python
# Ejemplos de nombres v√°lidos
mi_nombre = "Juan"
edad_actual = 28
numero_telefono = "9991234567"
_privado = "secreto"

print(mi_nombre, edad_actual, numero_telefono)
````

## Reasignar Variables

Puedes cambiar el valor de una variable en cualquier momento:

````python
# Crear una variable
mensaje = "Hola"
print("Primer mensaje:", mensaje)

# Cambiar su valor
mensaje = "Adi√≥s"
print("Nuevo mensaje:", mensaje)
````

````python
# Ejemplo con n√∫meros
contador = 0
print("Contador inicial:", contador)

contador = 5
print("Contador actualizado:", contador)

contador = contador + 1
print("Contador incrementado:", contador)
````

## Operaciones B√°sicas con Variables

````python
# Operaciones matem√°ticas
a = 10
b = 3

suma = a + b
resta = a - b
multiplicacion = a * b
division = a / b

print(f"Suma: {suma}")
print(f"Resta: {resta}")
print(f"Multiplicaci√≥n: {multiplicacion}")
print(f"Divisi√≥n: {division}")
````

````python
# Concatenar (unir) textos
nombre = "Mar√≠a"
apellido = "Gonz√°lez"

nombre_completo = nombre + " " + apellido
print(nombre_completo)
````

## Variables con Entrada del Usuario

Podemos pedirle al usuario que ingrese informaci√≥n:

````python
# Pedir el nombre al usuario
tu_nombre = input("¬øC√≥mo te llamas? ")
print(f"¬°Hola, {tu_nombre}!")
````

````python
# Nota: input() siempre devuelve texto (string)
# Si necesitas un n√∫mero, debes convertirlo
edad_texto = input("¬øCu√°ntos a√±os tienes? ")
edad_numero = int(edad_texto)  # Convertir a n√∫mero entero

print(f"Tienes {edad_numero} a√±os")
print(f"El pr√≥ximo a√±o tendr√°s {edad_numero + 1} a√±os")
````
