## El Condicional if B√°sico

```python
# Ejemplo simple
edad = 20

if edad >= 18:
    print("Eres mayor de edad")
```

```python
# Otro ejemplo
temperatura = 30

if temperatura > 25:
    print("Hace calor")
    print("Lleva agua")
```

## if-else: Dos Caminos Posibles

```python
# Si la condici√≥n es True, ejecuta el if
# Si es False, ejecuta el else
edad = 15

if edad >= 18:
    print("Eres mayor de edad")
else:
    print("Eres menor de edad")
```

```python
# Ejemplo con n√∫meros
numero = -5

if numero > 0:
    print("El n√∫mero es positivo")
else:
    print("El n√∫mero es negativo o cero")
```

## if-elif-else: M√∫ltiples Opciones

```python
# elif = "else if" (sino, si...)
calificacion = 85

if calificacion >= 90:
    print("Excelente")
elif calificacion >= 80:
    print("Muy bien")
elif calificacion >= 70:
    print("Bien")
elif calificacion >= 60:
    print("Suficiente")
else:
    print("Necesitas mejorar")
```

```python
# Clasificar la edad
edad = 25

if edad < 13:
    categoria = "Ni√±o"
elif edad < 18:
    categoria = "Adolescente"
elif edad < 65:
    categoria = "Adulto"
else:
    categoria = "Adulto mayor"

print(f"Categor√≠a: {categoria}")
```

## Operadores de Comparaci√≥n

- `==` : Igual a
- `!=` : Diferente de
- `>` : Mayor que
- `<` : Menor que
- `>=` : Mayor o igual que
- `<=` : Menor o igual que

```python
# Ejemplos de comparaciones
x = 10
y = 20

print(f"{x} == {y}:", x == y)
print(f"{x} != {y}:", x != y)
print(f"{x} > {y}:", x > y)
print(f"{x} < {y}:", x < y)
print(f"{x} >= {y}:", x >= y)
print(f"{x} <= {y}:", x <= y)
```

```python
# Comparar strings
nombre = "Ana"

if nombre == "Ana":
    print("¬°Hola Ana!")
else:
    print("No eres Ana")
```

## Operadores L√≥gicos

- `and` : Ambas condiciones deben ser True
- `or` : Al menos una condici√≥n debe ser True
- `not` : Invierte el valor (True se vuelve False y viceversa)

```python
# Operador and
edad = 25
tiene_licencia = True

if edad >= 18 and tiene_licencia:
    print("Puede conducir")
else:
    print("No puede conducir")
```

```python
# Operador or
es_fin_semana = True
es_feriado = False

if es_fin_semana or es_feriado:
    print("¬°Puedes descansar!")
else:
    print("D√≠a laboral")
```

```python
# Operador not
esta_lloviendo = False

if not esta_lloviendo:
    print("Puedes salir sin paraguas")
else:
    print("Lleva paraguas")
```

```python
# Combinar operadores
edad = 20
es_estudiante = True
tiene_descuento = False

if (edad < 25 and es_estudiante) or tiene_descuento:
    print("Aplica para descuento especial")
else:
    print("Precio regular")
```

## Condicionales Anidados

```python
# Un if dentro de otro if
tiene_dinero = True
esta_abierto = True

if tiene_dinero:
    if esta_abierto:
        print("Puedes comprar")
    else:
        print("La tienda est√° cerrada")
else:
    print("No tienes dinero")
```

```python
# Ejemplo m√°s complejo
edad = 16
tiene_permiso = True

if edad >= 18:
    print("Mayor de edad - Puede entrar")
else:
    if tiene_permiso:
        print("Menor con permiso - Puede entrar")
    else:
        print("Menor sin permiso - No puede entrar")
```

## Ejemplos Pr√°cticos

```python
# Calculadora de IMC (√çndice de Masa Corporal)
peso = 70  # kg
altura = 1.75  # metros

imc = peso / (altura ** 2)
print(f"Tu IMC es: {imc:.2f}")

if imc < 18.5:
    print("Bajo peso")
elif imc < 25:
    print("Peso normal")
elif imc < 30:
    print("Sobrepeso")
else:
    print("Obesidad")
```

```python
# Sistema de descuentos
precio = 1000
cantidad = 5

total = precio * cantidad
descuento = 0

if cantidad >= 10:
    descuento = 0.20  # 20% de descuento
elif cantidad >= 5:
    descuento = 0.10  # 10% de descuento

total_con_descuento = total * (1 - descuento)

print(f"Subtotal: ${total}")
print(f"Descuento: {descuento * 100}%")
print(f"Total: ${total_con_descuento}")
```
