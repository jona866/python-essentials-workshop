# üìã Listas B√°sicas en Python

Las listas son una de las estructuras de datos m√°s √∫tiles en Python. Permiten almacenar m√∫ltiples valores en una sola variable.

## ¬øQu√© es una Lista?

Una lista es una colecci√≥n **ordenada** y **modificable** de elementos. Puede contener n√∫meros, textos, o incluso otras listas.

## Crear una Lista

Las listas se crean usando corchetes `[]`:

```python
# Lista vac√≠a
lista_vacia = []
print(lista_vacia)
```

```python
# Lista de n√∫meros
numeros = [1, 2, 3, 4, 5]
print(numeros)
```

```python
# Lista de nombres
nombres = ["Ana", "Juan", "Mar√≠a", "Carlos"]
print(nombres)
```

```python
# Lista con diferentes tipos de datos
mixta = ["Python", 3, 3.14, True]
print(mixta)
```

## Acceder a Elementos de una Lista

Cada elemento de una lista tiene una posici√≥n (√≠ndice) que empieza en **0**.

```python
frutas = ["manzana", "pl√°tano", "naranja", "uva"]

# Acceder al primer elemento (√≠ndice 0)
print(frutas[0])  # manzana

# Acceder al segundo elemento (√≠ndice 1)
print(frutas[1])  # pl√°tano

# Acceder al √∫ltimo elemento
print(frutas[3])  # uva
```

```python
# √çndices negativos: acceden desde el final
print(frutas[-1])  # √∫ltimo elemento: uva
print(frutas[-2])  # pen√∫ltimo elemento: naranja
```

## Longitud de una Lista

Usa `len()` para saber cu√°ntos elementos tiene una lista:

```python
numeros = [10, 20, 30, 40, 50]
print(f"La lista tiene {len(numeros)} elementos")

nombres = ["Ana", "Juan", "Mar√≠a"]
print(f"La lista de nombres tiene {len(nombres)} elementos")
```

## Agregar Elementos: append()

El m√©todo `append()` a√±ade un elemento al final de la lista:

```python
# Crear una lista
colores = ["rojo", "azul"]
print("Lista original:", colores)

# Agregar un elemento
colores.append("verde")
print("Despu√©s de agregar:", colores)

# Agregar otro
colores.append("amarillo")
print("Despu√©s de agregar otro:", colores)
```

```python
# Ejemplo pr√°ctico: construir una lista desde cero
numeros = []
print("Lista vac√≠a:", numeros)

numeros.append(10)
numeros.append(20)
numeros.append(30)
print("Lista con elementos:", numeros)
```

## Modificar Elementos

Puedes cambiar el valor de un elemento usando su √≠ndice:

```python
animales = ["perro", "gato", "pez"]
print("Lista original:", animales)

# Cambiar el segundo elemento
animales[1] = "hamster"
print("Lista modificada:", animales)
```

## Eliminar Elementos

Hay varias formas de eliminar elementos:

```python
# Con remove(): elimina la primera aparici√≥n del valor
numeros = [10, 20, 30, 20, 40]
print("Lista original:", numeros)

numeros.remove(20)
print("Despu√©s de remove(20):", numeros)
```

```python
# Con pop(): elimina y devuelve el √∫ltimo elemento
frutas = ["manzana", "pl√°tano", "naranja"]
print("Lista original:", frutas)

ultima = frutas.pop()
print("Elemento eliminado:", ultima)
print("Lista despu√©s de pop():", frutas)
```

```python
# pop() con √≠ndice: elimina el elemento en esa posici√≥n
colores = ["rojo", "azul", "verde", "amarillo"]
print("Lista original:", colores)

eliminado = colores.pop(1)  # elimina el √≠ndice 1 (azul)
print("Elemento eliminado:", eliminado)
print("Lista resultante:", colores)
```

## Verificar si un Elemento Est√° en la Lista

Usa `in` para comprobar si un valor existe en la lista:

```python
frutas = ["manzana", "pl√°tano", "naranja"]

# Verificar si existe
print("manzana" in frutas)  # True
print("uva" in frutas)      # False
```

```python
# Uso con condicionales
if "pl√°tano" in frutas:
    print("¬°Tenemos pl√°tanos!")
else:
    print("No hay pl√°tanos")
```

## Recorrer una Lista

Puedes usar un ciclo `for` para iterar sobre cada elemento:

```python
# Recorrer e imprimir cada elemento
nombres = ["Ana", "Juan", "Mar√≠a", "Carlos"]

for nombre in nombres:
    print(f"Hola, {nombre}!")
```

```python
# Recorrer con √≠ndice
frutas = ["manzana", "pl√°tano", "naranja"]

for i in range(len(frutas)):
    print(f"Fruta {i}: {frutas[i]}")
```

## Listas y entrada del usuario

Puedes crear listas basadas en la entrada del usuario:

```python
# Crear una lista de tareas
tareas = []

tarea1 = input("Ingresa tu primera tarea: ")
tareas.append(tarea1)

tarea2 = input("Ingresa tu segunda tarea: ")
tareas.append(tarea2)

tarea3 = input("Ingresa tu tercera tarea: ")
tareas.append(tarea3)

print("\nTus tareas son:")
for tarea in tareas:
    print(f"- {tarea}")
```

## M√©todos √ötiles de Listas

Resumen de m√©todos comunes:

```python
numeros = [3, 1, 4, 1, 5, 9, 2, 6]

# Ordenar la lista
numeros.sort()
print("Lista ordenada:", numeros)

# Invertir el orden
numeros.reverse()
print("Lista invertida:", numeros)

# Contar cu√°ntas veces aparece un elemento
print("El 1 aparece", numeros.count(1), "veces")

# Encontrar el √≠ndice de un elemento
print("El 5 est√° en el √≠ndice:", numeros.index(5))
```

## üëÄ Vista r√°pida: Tuplas y Diccionarios

Adem√°s de las listas, Python tiene otras estructuras muy usadas:

- **Tuplas (`tuple`)**: parecidas a las listas, pero **no se pueden modificar** (son inmutables).
- **Diccionarios (`dict`)**: guardan pares **clave ‚Üí valor**, como una agenda de tel√©fonos.

En esta sesi√≥n solo queremos que **te suenen estos conceptos**. Los veremos con m√°s detalle m√°s adelante.

```python
# Ejemplo de tupla: parecida a una lista, pero no se puede modificar
coordenadas = (10, 20)
print(coordenadas)
print(type(coordenadas))

# Podemos leer sus valores por √≠ndice
x = coordenadas[0]
y = coordenadas[1]
print("x =", x, "y =", y)

# Si intentas cambiar un valor, obtendr√°s un error (pru√©balo m√°s tarde):
# coordenadas[0] = 50  # TypeError: 'tuple' object does not support item assignment
```

```python
# Ejemplo de diccionario: pares clave ‚Üí valor
persona = {
    "nombre": "Ana",
    "edad": 28,
    "ciudad": "M√©rida"
}

print(persona)
print(type(persona))

# Acceder a un valor usando su clave
print("Nombre:", persona["nombre"])
print("Edad:", persona["edad"])

# Agregar una nueva clave
persona["profesion"] = "Desarrolladora"
print(persona)

# No te preocupes si esto parece nuevo: lo veremos con calma en sesiones futuras.
```
