## Funciones con Par√°metros

Los **par√°metros** son valores que la funci√≥n recibe para trabajar con ellos.

````python
# Funci√≥n con un par√°metro
def saludar_persona(nombre):
    print(f"¬°Hola, {nombre}!")

# Llamar con diferentes nombres
saludar_persona("Ana")
saludar_persona("Carlos")
saludar_persona("Mar√≠a")
````

````python
# Funci√≥n con m√∫ltiples par√°metros
def presentar(nombre, edad, ciudad):
    print(f"Me llamo {nombre}, tengo {edad} a√±os y vivo en {ciudad}")

presentar("Juan", 25, "M√©rida")
presentar("Laura", 30, "Canc√∫n")
````

## Funciones que Retornan Valores

Una funci√≥n puede **devolver** un valor usando `return`.

````python
# Funci√≥n que suma dos n√∫meros
def sumar(a, b):
    resultado = a + b
    return resultado

# Usar el valor retornado
total = sumar(5, 3)
print(f"5 + 3 = {total}")

# Tambi√©n podemos usarlo directamente
print(f"10 + 20 = {sumar(10, 20)}")
````

````python
# Calculadora b√°sica con funciones
def sumar(a, b):
    return a + b

def restar(a, b):
    return a - b

def multiplicar(a, b):
    return a * b

def dividir(a, b):
    if b == 0:
        return "Error: No se puede dividir entre cero"
    return a / b

# Usar las funciones
print(f"5 + 3 = {sumar(5, 3)}")
print(f"10 - 4 = {restar(10, 4)}")
print(f"6 * 7 = {multiplicar(6, 7)}")
print(f"15 / 3 = {dividir(15, 3)}")
print(f"10 / 0 = {dividir(10, 0)}")
````

## Par√°metros con Valores por Defecto

````python
# Si no se proporciona un valor, usa el valor por defecto
def saludar(nombre, mensaje="Bienvenido"):
    print(f"{mensaje}, {nombre}!")

# Con el valor por defecto
saludar("Ana")

# Con un mensaje personalizado
saludar("Carlos", "Hola")
saludar("Mar√≠a", "Buenos d√≠as")
````

````python
# Calcular √°rea de rect√°ngulo
def area_rectangulo(base, altura=1):
    return base * altura

print(f"√Årea (5 x 3): {area_rectangulo(5, 3)}")
print(f"√Årea (10 x 1): {area_rectangulo(10)}")  # altura = 1 por defecto
````

## Retornar M√∫ltiples Valores

````python
# Una funci√≥n puede retornar varios valores
def operaciones_basicas(a, b):
    suma = a + b
    resta = a - b
    multiplicacion = a * b
    return suma, resta, multiplicacion

# Recibir los valores
s, r, m = operaciones_basicas(10, 5)
print(f"Suma: {s}")
print(f"Resta: {r}")
print(f"Multiplicaci√≥n: {m}")
````

````python
# Calcular estad√≠sticas de una lista
def estadisticas(numeros):
    suma = sum(numeros)
    promedio = suma / len(numeros)
    maximo = max(numeros)
    minimo = min(numeros)
    return suma, promedio, maximo, minimo

datos = [10, 25, 18, 30, 12]
s, p, max_val, min_val = estadisticas(datos)

print(f"Suma: {s}")
print(f"Promedio: {p}")
print(f"M√°ximo: {max_val}")
print(f"M√≠nimo: {min_val}")
````

## √Åmbito de Variables (Scope)

Las variables dentro de una funci√≥n son **locales** (solo existen dentro de la funci√≥n).

````python
# Variable global
mensaje_global = "Soy global"

def mi_funcion():
    # Variable local
    mensaje_local = "Soy local"
    print(mensaje_global)  # Puedo acceder a la global
    print(mensaje_local)   # Puedo acceder a la local

mi_funcion()
print(mensaje_global)  # Funciona
# print(mensaje_local)  # Error: no existe fuera de la funci√≥n
````

## Funciones que Llaman a Otras Funciones

````python
def calcular_area_circulo(radio):
    pi = 3.14159
    return pi * (radio ** 2)

def calcular_area_cilindro(radio, altura):
    area_base = calcular_area_circulo(radio)
    area_lateral = 2 * 3.14159 * radio * altura
    area_total = (2 * area_base) + area_lateral
    return area_total

radio = 5
altura = 10
area = calcular_area_cilindro(radio, altura)
print(f"√Årea del cilindro: {area:.2f}")
````

## Ejemplos Pr√°cticos

````python
# Validar si un n√∫mero es par
def es_par(numero):
    return numero % 2 == 0

print(f"¬ø4 es par? {es_par(4)}")
print(f"¬ø7 es par? {es_par(7)}")
````

````python
# Convertir temperatura
def celsius_a_fahrenheit(celsius):
    return (celsius * 9/5) + 32

def fahrenheit_a_celsius(fahrenheit):
    return (fahrenheit - 32) * 5/9

temp_c = 25
temp_f = celsius_a_fahrenheit(temp_c)
print(f"{temp_c}¬∞C = {temp_f}¬∞F")

temp_f2 = 77
temp_c2 = fahrenheit_a_celsius(temp_f2)
print(f"{temp_f2}¬∞F = {temp_c2:.1f}¬∞C")
````

````python
# Calcular precio con descuento
def aplicar_descuento(precio, descuento_porcentaje):
    descuento = precio * (descuento_porcentaje / 100)
    precio_final = precio - descuento
    return precio_final, descuento

precio_original = 1000
precio_final, ahorro = aplicar_descuento(precio_original, 20)

print(f"Precio original: ${precio_original}")
print(f"Descuento (20%): ${ahorro}")
print(f"Precio final: ${precio_final}")
````

````python
# Generar una contrase√±a simple
def generar_password(nombre, edad):
    # Tomar las primeras 3 letras del nombre y agregar la edad
    password = nombre[:3].lower() + str(edad) + "!"
    return password

nombre_usuario = "Carlos"
edad_usuario = 25
password = generar_password(nombre_usuario, edad_usuario)
print(f"Password generada: {password}")
````

## Documentar Funciones (Docstrings)

````python
def calcular_imc(peso, altura):
    """
    Calcula el √çndice de Masa Corporal (IMC).
    
    Par√°metros:
        peso (float): Peso en kilogramos
        altura (float): Altura en metros
    
    Retorna:
        float: El IMC calculado
    """
    imc = peso / (altura ** 2)
    return imc

# Usar la funci√≥n
mi_imc = calcular_imc(70, 1.75)
print(f"Tu IMC es: {mi_imc:.2f}")

# Ver la documentaci√≥n
help(calcular_imc)
````
