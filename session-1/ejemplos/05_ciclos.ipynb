## El Ciclo for B√°sico

```python
# Imprimir del 1 al 5
for i in range(1, 6):
    print(i)
```

```python
# Repetir un mensaje 5 veces
for i in range(5):
    print(f"Esta es la repetici√≥n n√∫mero {i + 1}")
```

## La Funci√≥n range()

`range()` genera una secuencia de n√∫meros:
- `range(5)`: 0, 1, 2, 3, 4 (5 n√∫meros desde 0)
- `range(1, 6)`: 1, 2, 3, 4, 5 (desde 1 hasta 5)
- `range(0, 10, 2)`: 0, 2, 4, 6, 8 (de 2 en 2)

```python
# range(inicio, fin)
print("N√∫meros del 1 al 10:")
for num in range(1, 11):
    print(num, end=" ")
print()  # Salto de l√≠nea
```

```python
# range(inicio, fin, paso)
print("N√∫meros pares del 0 al 20:")
for num in range(0, 21, 2):
    print(num, end=" ")
print()
```

```python
# Contar hacia atr√°s
print("Cuenta regresiva:")
for num in range(10, 0, -1):
    print(num, end=" ")
print("¬°Despegue!")
```

## Iterar sobre Listas

```python
# Iterar sobre una lista de nombres
nombres = ["Ana", "Carlos", "Mar√≠a", "Juan", "Laura"]

for nombre in nombres:
    print(f"Hola, {nombre}!")
```

```python
# Iterar sobre una lista de n√∫meros
precios = [10.50, 25.00, 15.75, 8.99]

total = 0
for precio in precios:
    total = total + precio

print(f"Total: ${total}")
```

```python
# Iterar con √≠ndice usando enumerate()
frutas = ["manzana", "banana", "naranja", "uva"]

for indice, fruta in enumerate(frutas):
    print(f"{indice + 1}. {fruta}")
```

## El Ciclo while

El ciclo `while` repite mientras una condici√≥n sea verdadera.

```python
# Contar del 1 al 5 con while
contador = 1

while contador <= 5:
    print(contador)
    contador = contador + 1
```

```python
# Sumar n√∫meros hasta llegar a 100
suma = 0
numero = 1

while suma < 100:
    suma = suma + numero
    numero = numero + 1

print(f"Se necesitaron {numero - 1} n√∫meros para llegar a {suma}")
```

## break: Salir del Ciclo

```python
# Buscar un n√∫mero en una lista
numeros = [3, 7, 12, 18, 25, 30]
buscar = 18

for num in numeros:
    if num == buscar:
        print(f"¬°Encontrado! El n√∫mero {buscar} est√° en la lista")
        break
    print(f"Revisando {num}...")
```

```python
# Salir cuando se cumpla una condici√≥n
intentos = 0
max_intentos = 5

while True:
    intentos += 1
    print(f"Intento {intentos}")
    
    if intentos >= max_intentos:
        print("M√°ximo de intentos alcanzado")
        break
```

## continue: Saltar a la Siguiente Iteraci√≥n

```python
# Imprimir solo n√∫meros impares
for num in range(1, 11):
    if num % 2 == 0:  # Si es par
        continue  # Saltar al siguiente n√∫mero
    print(num)
```

```python
# Procesar solo valores v√°lidos
valores = [10, -5, 15, 0, 20, -3, 8]

suma_positivos = 0
for valor in valores:
    if valor <= 0:
        continue  # Ignorar valores no positivos
    suma_positivos += valor

print(f"Suma de valores positivos: {suma_positivos}")
```

## Ciclos Anidados

```python
# Tabla de multiplicar
for i in range(1, 4):
    print(f"\nTabla del {i}:")
    for j in range(1, 6):
        resultado = i * j
        print(f"{i} x {j} = {resultado}")
```

```python
# Patr√≥n de asteriscos
filas = 5

for i in range(1, filas + 1):
    for j in range(i):
        print("*", end="")
    print()  # Nueva l√≠nea
```

## Ejemplos Pr√°cticos

```python
# Calcular factorial
numero = 5
factorial = 1

for i in range(1, numero + 1):
    factorial = factorial * i

print(f"El factorial de {numero} es {factorial}")
```

```python
# Validar entrada del usuario
intentos = 0
max_intentos = 3
password_correcto = "python123"

while intentos < max_intentos:
    password = input("Ingresa la contrase√±a: ")
    
    if password == password_correcto:
        print("¬°Acceso concedido!")
        break
    else:
        intentos += 1
        restantes = max_intentos - intentos
        if restantes > 0:
            print(f"Contrase√±a incorrecta. Te quedan {restantes} intentos")
        else:
            print("Acceso bloqueado")
```

```python
# Generar serie de Fibonacci
n = 10
a, b = 0, 1

print(f"Primeros {n} n√∫meros de Fibonacci:")
for i in range(n):
    print(a, end=" ")
    a, b = b, a + b
print()
```
